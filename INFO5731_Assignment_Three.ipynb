```python
# Write your code here
import nltk
import pandas as pd
df = pd.read_csv('shangchi_movieR.csv')
df['word_count'] = df['movie_Reviews'].apply(lambda x:len(str(x).split(" ")))
df

```

```python
import nltk
nltk.download()
```

```python
df['movie_Reviews'] = df['movie_Reviews'].str.replace('[^\w\s]','')  #removal of punctuation
df['movie_Reviews'] = df['movie_Reviews'].apply(lambda x: " ".join(x.lower() for x in x.split()))#lower case
from nltk.corpus import stopwords #removal of stopwords
nltk.download('stopwords')
stop = stopwords.words('english')
df['movie_Reviews'] = df['movie_Reviews'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
```

```python
#changing to a single line file
line = ''
for i in range(0,len(df['movie_Reviews'])):
  line+= df['movie_Reviews'][i]
print(line)
```

 **1.Count the frequency of all the N-grams (N=3).**

```python
#trigrams
nltk.download('punkt')
words = nltk.word_tokenize(line)
result_trigrams = nltk.trigrams(words)
freq_trigrams = nltk.FreqDist(result_trigrams)
freq_trigrams
```

```python
#bigrams
result_bigrams = nltk.bigrams(words)
import re
from itertools import tee, islice
import collections
def ngrams(lst, n):
  tlst = lst
  while True:
    a, b = tee(tlst)
    l = tuple(islice(a, n))
    if len(l) == n:
      yield l
      next(b)
      tlst = b
    else:
      break
line2 = collections.Counter(ngrams(line.split(), 2))
print(line2)
```

```python
s1 = ''
for w1,w2 in result_bigrams:
  s1 = s1 + ' ' + w1 + ' ' + w2
list = s1.split()
print(list)
```

**2) Calculate the probabilities for all the bigrams in the dataset by using the fomular count(w2 w1) / count(w2). For example, count(really like) / count(really) = 1 / 3 = 0.33.**

```python
for w1,w2 in line2:  
  den = list.count(w1)
  num = line2[(w1,w2)]
  if den == 0:
    break
  prob = num/den
  print(w1,w2,prob)
```

**(3) Extract all the noun phrases and calculate the relative probabilities of each review in terms of other reviews (abstracts, or tweets) by using the fomular frequency (noun phrase) / max frequency (noun phrase) on the whole dataset. Print out the result in a table with column name the all the noun phrases and row name as the 500 reviews (abstracts, or tweets).**

```python
from collections import Counter      
import nltk
from nltk import word_tokenize, pos_tag, pos_tag_sents
nltk.download('averaged_perceptron_tagger')
lower_case = line.lower()
tokens = nltk.word_tokenize(lower_case)
tags = nltk.pos_tag(tokens)
counts = Counter( tag for word,  tag in tags)
highest_noun_count = counts['NN']
df['phrases'] = pos_tag_sents(df['movie_Reviews'].apply(word_tokenize).tolist())
y = df['phrases']
list1 = []
list2 = []
for i in y:
  count = Counter( tag for word,  tag in i)
  list1.append(count['NN'])
for i in list1:
  j = i/highest_noun_count
  list2.append(j)
print(list2)
df1 = pd.DataFrame(list1,columns = ['noun_count'])
df1
df2 = pd. DataFrame(list2,columns = ['prob_noun'])
result = pd.concat([df1,df2], axis=1, sort=False)
result
df = pd.concat([df,result], axis = 1, sort = False)
#del df['noun_phrases']
df
```

**(2) To rank the documents with respect to query (design a query by yourself, for example, "An Outstanding movie with a haunting performance and best character development") by using cosine similarity.**

```python
#cleaning
import nltk  
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
query = "The movie is pretty good in most ways, but Akwafina's character is just... annoying. Sorry I didn't like her in this"
lemmatizer = WordNetLemmatizer()
regex = re.compile('[^a-zA-Z\s]')

def cleaningdata(text):
  text =  re.sub(regex, '', text) # Removes Special characters 
  words = " ".join([word for word in text.split() if word not in stopwords.words('english')]) # removes stopwords from text
  lemmatize =   " ".join([lemmatizer.lemmatize(i) for i in words.lower().split()]) # makes word in to meaning
  return lemmatize
cleaned_query = cleaningdata(query)
print('Cleaned Query: ' + cleaned_query)
```

```python
#cos-sim
import pandas
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
from nltk import corpus
text = []
for x in df['movie_Reviews']:
  text.append(x)

text.append(cleaned_query)
vectorizer = TfidfVectorizer(input=text, min_df=0.1, stop_words='english')
matrix = vectorizer.fit_transform(text)

cs_dict = {}
for i in range(0, matrix.shape[0]):
  sim=cosine_similarity(matrix[80], matrix[i])
  cs_dict[str(i)] = sim[0][0]
  i+=1

sorted_keys = (sorted(cs_dict, key=cs_dict.get,reverse=True))
sorted_cosine_dict = {}
for key in sorted_keys:
  sorted_cosine_dict[key] = cs_dict[key];

movie_Reviews = []
cos = []
for y in sorted_cosine_dict:
  movie_Reviews.append([int(y)])
  cos.append(sorted_cosine_dict[y])
reviewcos = pd.DataFrame(columns={'Reviews'})
reviewcos['Reviews'] = movie_Reviews
reviewcos['Rank'] = [i for i in range(1, len(sorted_keys)+1)]
reviewcos['Cosinesimilarity'] = cos
reviewcos
```
